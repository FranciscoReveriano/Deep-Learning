## Problem 1 : Exponential Distribution

I first declare the Exponential Function. I plot this on Matplotlib. Then to check I plot it again on SNS.

### Starting Functions

```python
# Exponential Distribution Function
def exponential_distribution(chosen_lambda, number):
    return (-1/chosen_lambda) * math.log(1 - number)


# Exponential PDF Function
def exponential_density_pdf(x, chosen_lambda):
    return chosen_lambda * math.exp(-chosen_lambda * x)
    
```

## Problem 2: Gamma Distribution

### Starting Functions

```python
# First I need to Define the Factorial Calculation 
def factorial(n):
    if n==1:
        return n;
    elif n<1:
        return ("Error")
    else:
        return n*factorial(n-1)

def gamma(n):
    return math.factorial(n-1)
```

```python
# Function to Calculate Lambda
def calculate_lambda(beta):
    chosen_lambda = 1 / beta
    return chosen_lambda

# The Gamma Distribution calls a Random Uniform Value When Calculating the Exponential Distribution
def gamma_distribution(K, Beta):
    # Proceed to Calculate the Correct Lambda
    chosen_lambda = calculate_lambda(Beta)
    # Summation Dummy Variable
    summation = 0
    # Proceed to Loop Through To Calculate the Gamma Random Value 
    for i in range(K):
        summation += exponential_distribution(chosen_lambda, random.random())
    return summation

# The Gamma Distribution PDF
def gamma_distribution_pdf(x, K, B):
    chosen_lambda = calculate_lambda(B)
    numerator = (x**(K-1))*(chosen_lambda**K)
    denominator = gamma(K)
    return numerator/denominator * math.exp(-chosen_lambda * x)
```

### Gamma Distribution = 0.1

#### Plotting PDF

```python
Beta = 0.1
K = 5
samples = 1 * 10**5

# Histogram of Variates
gamma_distribution_list = []
for i in range(samples):
    answer = gamma_distribution(K, Beta)
    gamma_distribution_list.append(answer)
    

# Calculating the PDF
gamma_pdf_x = np.arange(0,4,0.1)
gamma_pdf_y = []
for num in gamma_pdf_x:
    y = gamma_distribution_pdf(num, K, Beta)
    gamma_pdf_y.append(y)
    
    
## Graph PDF 
plt.plot(gamma_pdf_x, gamma_pdf_y)
plt.title("Gamma PDF")
plt.show()
```

## Problem 3: Beta Distribution

### Starting Functions

```python
import scipy
```

```python
def combination(n,k):
    return factorial(n) / (factorial(k) * factorial(n-k))
```

```python
# Beta Distribution Probability Density Function
def beta_distribution_pdf(x,a,b):
    numerator = gamma(a + b)*(x**(a-1))*(1-x)**(b-1)
    denominator = gamma(a) * gamma(b)
    return numerator/denominator


def beta_distribution2(a,b):
    k = a - 1
    n = b - 1 + k
    return (combination(n,k))*(x**(k))*((1-x)**(n-k)) * (n+1)
```

```python
def gamma_distributed_random_variable(n):
    summation = 0
    for i in range(n):
        summation += math.log(random.random())
    return summation

def beta_distribution(a,b):
    X = gamma_distributed_random_variable(a)
    Y = gamma_distributed_random_variable(b)
    return X / (X+Y)
```

## Problem 3: Dirichlet Distribution

### Using Gamma Method

```python
 def dirichlet_method1(N, alphas):
        solution = []
        gammas = []
        # Calculate the Sum in the Denominator
        denominator = 0
        for i in range(N):
            gamma_i= gamma_distribution(alphas[i], 1)
            gammas.append(gamma_i)
            denominator += gamma_i
            
        # Calculate the Answers 
        for i in range(N):
            answer_y = gammas[i]/denominator
            solution.append(answer_y)
        
        return solution
```

```python
N = 3
Alphas = [10,10,10]
samples = 1 * 10**5
dirichlet_distribution_list = []
first_feature = []
second_feature = []
for i in range(samples):
    answer = dirichlet_method1(N, Alphas)
    dirichlet_distribution_list.append(answer)
    first_feature.append(answer[0])
    second_feature.append(answer[1])

binwidth = 0.01
bins = np.arange(min(first_feature), max(second_feature), binwidth)
plt.hist2d(first_feature, second_feature, density=True, bins=bins)
plt.colorbar()
plt.show()
```

### Using Beta Stick Method

```python
def dirichlet_beta_stick(N, alphas):
    ys = []
    xs = []
    # Calculate Xj
    for i in range(N-1):
        xj = beta_distribution(alphas[i], np.sum(alphas[i+1:]))
        xs.append(xj)
    for i in range(N-1):
        if i == 0:
            ys.append(xs[0])
        else:
            answer = (1- xs[i-1])*xs[i]
            ys.append(answer)
    return ys
```

```python
N = 3
Alphas = [10,10,10]
samples = 1 * 10**5
dirichlet_distribution_list = []
first_feature = []
second_feature = []
for i in range(samples):
    answer = dirichlet_beta_stick(N, Alphas)
    dirichlet_distribution_list.append(answer)
    first_feature.append(answer[0])
    second_feature.append(answer[1])

binwidth = 0.01
bins = np.arange(min(first_feature), max(second_feature), binwidth)
plt.hist2d(first_feature, second_feature, density=True, bins=bins)
plt.colorbar()
plt.show()
```
