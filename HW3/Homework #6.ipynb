### Parellel Distribution

```python
def parallel(t):
    answer = []
    for i in range(len(t)):
        sum_value = 1
        results = 0
        for j in range(len(t[i])):
            sum_value *= (1 - t[i][j])
        results = 1 - sum_value
        answer.append(results)
    return answer
```

```python
m = 1000

y1j = []
min_list = []
for i in range(m):
    list1 = []
    #random.seed(7)
    x11 = exp(random.random())
    # Second Random Number
    #random.seed(10)
    x21 = exp(random.random())
    # Third Random Number
    #random.seed(13)
    x31 = exp(random.random())
    
    # Append to List
    list1.append(x11)
    list1.append(x21)
    list1.append(x31)
    
    # Append Tuple List
    y1j.append(list1)
    
    # Append Tuple List
    tup1 = (x11, x21, x31)
    min1 = max(tup1)
    min_list.append(min1)

y1j = parallel(y1j)


plt.figure(figsize=(10,5))
plt.subplot(131)
sns.distplot(y1j,  bins=10)
plt.title("Parallel : m=1000")
plt.subplot(122)
sns.distplot(min_list, bins=10)
plt.title("Maximum Distribution : m=1000")
```

### K of N System

```python
def binom(n, k):
    return math.factorial(n) // math.factorial(k) // math.factorial(n - k)
```

```python
def kofn(k, n, r):
    sum_value = 0
    n = n + 1
    for j in range(k,n):
        sum_value += binom(n,j) * exp(r)**j * (1-exp(r))**(n-j)
    return sum_value
        
```

```python
m = 1000

y1j = []
min_list = []
for i in range(m):
    list1 = []
    #random.seed(7)
    num1 = random.random()
    x11 = exp(num1)
    # Second Random Number
    #random.seed(10)
    x21 = exp(random.random())
    # Third Random Number
    #random.seed(13)
    x31 = exp(random.random())
    
    # Append to List
    response = kofn(2,3,num1)
    
    # Append Tuple List
    y1j.append(response)
    
    # Append Tuple List
    tup1 = (x11, x21, x31)
    min1 = median(tup1)
    min_list.append(min1)


# Proceed to calculate the line
x_array = np.arange(0,1,0.01)
y_array = np.zeros(x_array.shape)
for i in range(len(x_array)):
    y_array[i] = kofn(2,3,x_array[i])
    
plt.figure(figsize=(10,5))
plt.subplot(131)
sns.distplot(y1j,  bins=10)
plt.title("KofN : m= 1000")
plt.subplot(132)
sns.distplot(min_list, bins=10)
plt.title("Median Distribution : m=1000")
plt.subplot(133)
plt.title("Attempt to plot distribution")
plt.plot(x_array, y_array)
```

This is very interesting because we can see that the largest histogram is still the same one. Meaning that the value is consistently falling on the same values. Showing that it does a good job predicitng.

## Plotting TMR

```python
def TMR(t):
    return 3*exp(2*t) - 2*exp(3*t)
```

```python
m = 1000
x_array = np.arange(0,1,0.001)
y_array = np.zeros(x_array.shape)
for i in range(m):
    x_array[i] = random.random()
    y_array[i] = TMR(x_array[i])
    
x_array1 = np.arange(0,1,0.001)
y_array1 = np.zeros(x_array.shape)
for i in range(m):
    #x_array[i] = random.random()
    y_array1[i] = exp(x_array1[i])

plt.figure(figsize=(10,5))
plt.scatter(x_array, y_array, color="blue", label="TMR")
plt.scatter(x_array1, y_array1, color="green", label="Single Component")
plt.legend()
plt.title("TMR Reliability")
```
